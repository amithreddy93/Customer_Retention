```python
import matplotlib.pyplot as plt
import seaborn as sns
for i, col in enumerate(df_website.columns):
    col = df_website[col].str.split(',',expand=True).stack().tolist()
    plt.figure(i)
    plt.title(df_website.columns.tolist()[i])
    chart = sns.countplot(x =col,data=df_website)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
```

# Gratification: It is the degree to which an e-store is entertaining and fun to the online customers, it is also the feeling of accomplishment after completing a task easily on the website. Other form of gratification includes receiving incentives for shopping on a particular website.

```python
fig_dims = (10, 6)
fig, ax = plt.subplots(figsize=fig_dims)
a = sns.countplot(x='45 You feel gratification shopping on your favorite e-tailer',hue='6 How many times you have made an online purchase in the past 1 year?',data=df_main,ax=ax)
plt.legend(loc='upper right')
a.set_xticklabels(a.get_xticklabels(), rotation=90)
```

# Role: Role shopping refers to the online customers who shop to gain pleasure and self-approval when shopping under a different identity to perform a specific role at given time. It is the enjoyment that the online shoppers gets from shopping for the consumption of others (for example, father shopping for a child), shopping online for others, has been revealed as a source of joy for some customers.

```python
%matplotlib inline
g = sns.catplot(x="46 Shopping on the website helps you fulfill certain roles", hue="6 How many times you have made an online purchase in the past 1 year?", col="5 Since How Long You are Shopping Online ?",col_wrap=2, data=df_main, kind="count", height=5, aspect=1.5,legend_out=False)
g.set_xticklabels(rotation=90)

plt.show()
```
